```python
from scipy.stats import zscore

# Compute Z-scores
z_scores = df_continuous.iloc[:, :-1].apply(zscore)

# Identify outliers (Z-score threshold = ±3)
outliers_z = (z_scores.abs() > 3).sum()
print(outliers_z)
```

#### Even with the zscore method, we find no outliers

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df):
    df = df.iloc[:, :-1]
    outlier_dict = {}
    for col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outlier_dict[col] = outliers

    return outlier_dict

outliers_iqr = detect_outliers_iqr(df_continuous)
print(outliers_iqr)
```

#### Now we shall proceed for Feature Selection using Lasso regularization. If the feature is irrelevant, lasso penalizes it’s coefficient and make it 0. Hence the features with coefficient = 0 are removed and the rest are taken.

```python
from sklearn.linear_model import LassoCV

# Step 1: Fit LassoCV
reg = LassoCV(cv=10, random_state=42)  # cv is the number of folds for cross-validation
reg.fit(X, y)

# Step 2: Extract the best alpha (regularization strength) and model score
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" % reg.score(X, y))

# Step 3: Get coefficients and filter features
coef = pd.Series(reg.coef_, index=X.columns)
print("Lasso picked " + str(sum(coef != 0)) + " features and eliminated the other " + str(sum(coef == 0)) + " features")
```

#### Let us see with another technique: Recursive Feature Elimination (RFE) which is a feature selection technique that recursively removes the least important features and builds a model on the remaining features. It ranks features based on their importance using a chosen model and accuracy metric, where features with higher importance are retained. The optimal number of features is determined by evaluating the model's performance as features are removed.

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split

# Step 1: Loop to find the optimal number of features
nof_list = np.arange(1, X.shape[1] + 1)  # Number of features from 1 to total features
high_score = 0
nof = 0
score_list = []

for n in nof_list:
    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    #print("no. of iteration ", n)
    # Step 2: Initialize Linear Regression and RFE
    model = LinearRegression()
    rfe = RFE(model, n_features_to_select=n)  # Set number of features for RFE
    X_train_rfe = rfe.fit_transform(X_train, y_train)
    X_test_rfe = rfe.transform(X_test)

    # Fit the model and calculate the score
    model.fit(X_train_rfe, y_train)
    score = model.score(X_test_rfe, y_test)  # R^2 score

    score_list.append(score)

    if score > high_score:
        high_score = score
        nof = n  # Update the optimal number of features

print("Optimum number of features: %d" % nof)
print("Score with %d features: %f" % (nof, high_score))

# Step 3: Final RFE with optimum number of features
model_f = LinearRegression()
rfe_f = RFE(model_f, n_features_to_select=nof)
X_rfe = rfe_f.fit_transform(X, y)  # Fit on all data

# Step 4: Get the selected features
cols = list(X.columns)
temp = pd.Series(rfe_f.support_, index=cols)
selected_features_rfe = temp[temp == True].index

print("Features Selected by RFE:")
print(selected_features_rfe)
```

#### So we see both the techniques suggest `fertilizer_used`, `irrigation_used`, `rainfall_mm`, `temperature_celsius`. And as Lasso Regularization suggested the `soil_type` feature we shall also take that into account for a comprehensive Feature Matrix

```python
final_selected_features = ['rainfall_mm', 'fertilizer_used', 'irrigation_used', 'temperature_celsius', 'soil_type=Clay', 'soil_type=Loam', 'soil_type=Sandy', 'soil_type=Chalky', 'soil_type=Peaty', 'soil_type=Silt']
feature_selected_X = X[final_selected_features]
```

## Model Training

#### Setting up the validation framework

```python
# prepare for the train test split by joining the processed feature matrix with the target feature
final_df = feature_selected_X.copy()
final_df['yield_tons_per_hectare'] = y.values
```

```python
final_df.shape
```

```python
# the actual train test split
final_X_full_train, final_X_test = train_test_split(final_df, test_size=0.2, random_state=42)
final_X_train, final_X_val = train_test_split(final_X_full_train, test_size=len(final_X_test), random_state=42)
```

```python
# verifying
final_X_full_train.shape
```

```python
# resetting the indices of the dataframes so each starts from 0
final_X_train = final_X_train.reset_index(drop=True)
final_X_val = final_X_val.reset_index(drop=True)
final_X_test = final_X_test.reset_index(drop=True)
```

```python
# setting the target feature segment aligned to each of the fearure matrix segment
y_train = final_X_train.yield_tons_per_hectare.values
y_val = final_X_val.yield_tons_per_hectare.values
y_test = final_X_test.yield_tons_per_hectare.values
```

```python
# removing the target feature segment from the feature matrix segment
del final_X_train['yield_tons_per_hectare']
del final_X_val['yield_tons_per_hectare']
del final_X_test['yield_tons_per_hectare']
```

```python
# verifying feature matrices and target features
final_X_train.shape, final_X_val.shape, final_X_test.shape, y_train.shape, y_val.shape, y_test.shape
```

#### Now let us start the training of the model with Linear Regression

```python
# Initialize the Linear Regression model
lr_model = LinearRegression()

# Train the model on the training data
lr_model.fit(final_X_train, y_train)
```

```python
y_pred_lr = lr_model.predict(final_X_val)
```

#### Check the metrics

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
def evaluateMetrics(y_true, y_pred):
  # Mean Squared Error
  mse_lr = mean_squared_error(y_true, y_pred)

  # Root Mean Squared Error
  rmse_lr = np.sqrt(mse_lr)

  # Mean Absolute Error
  mae_lr = mean_absolute_error(y_true, y_pred)

  # R-squared
  r2_lr = r2_score(y_true, y_pred)

  # Print Metrics
  print(f'MSE: {mse_lr:.4f}')
  print(f'RMSE: {rmse_lr:.4f}')
  print(f'MAE: {mae_lr:.4f}')
  print(f'R²: {r2_lr:.4f}')
```

```python
# evaluating and printing metrics for the Linear Regression model
evaluateMetrics(y_val, y_pred_lr)
```

#### We shall also do a Learning Curve (Train vs Validation MSE) which helps underfitting or overfitting.

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, val_scores = learning_curve(
    estimator=lr_model, X=final_X_train, y=y_train, scoring='neg_mean_squared_error', cv=5)

train_scores_mean = -np.mean(train_scores, axis=1)
val_scores_mean = -np.mean(val_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label="Training Error", marker='o')
plt.plot(train_sizes, val_scores_mean, label="Validation Error", marker='s')
plt.xlabel("Training Set Size")
plt.ylabel("Mean Squared Error")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

From the plot we can see As the training size increases, the training error slightly decreases and then continues to increase, indicating that the model becomes less specialized (less overfitting).

The validation error is consistently higher than the training error, but the gap between the two kept narrowing, which suggests good generalization.

The model performs consistently well across different training set sizes, meaning it is neither underfitting nor overfitting.

#### We shall now train the model with Decision Tree Regressor

```python
# Import Necessary Library
from sklearn.tree import DecisionTreeRegressor
```

```python
# finding the best max_depth value

# Store results
max_depth_list = np.arange(1, 21)
r2_scores = []
rmse_scores = []
for max_depth in max_depth_list:
    # Initialize the Decision Tree Regressor
    dt_model = DecisionTreeRegressor(random_state=42, max_depth=max_depth)

    # Train the model
    dt_model.fit(final_X_train, y_train)

    # Predict using the validation dataset
    y_pred_dt = dt_model.predict(final_X_val)

    # Compute R-squared score
    r2 = r2_score(y_val, y_pred_dt)
    r2_scores.append(r2)

    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_pred_dt))
    rmse_scores.append(rmse)

    print(f'max_depth = {max_depth} -------> R² = {r2:.4f}, RMSE = {rmse:.4f}')
    print()
```

```python
# Visualization
fig, ax1 = plt.subplots(figsize=(8, 5))

color = 'tab:blue'
ax1.set_xlabel('max-depth')
ax1.set_ylabel('R² Score', color=color)
ax1.plot(max_depth_list, r2_scores, marker='o', linestyle='-', color=color, label='R² Score')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('RMSE', color=color)
ax2.plot(max_depth_list, rmse_scores, marker='s', linestyle='--', color=color, label='RMSE')
ax2.tick_params(axis='y', labelcolor=color)

fig.suptitle('R² and RMSE vs. max_depth')
plt.show()
```

#### We can observe that best max_depth  = 10 from the plot

#### Now let us check the R² and RMSE for various min_samples_leaf values

```python
# Store results
leaf_samples_list = [1, 5, 10, 15, 20, 50, 100, 200, 500]
r2_scores = []
rmse_scores = []
for leaf_samples in leaf_samples_list:
    dt_model = DecisionTreeRegressor(random_state=42, max_depth=10, min_samples_leaf=leaf_samples)

    # Train the model
    dt_model.fit(final_X_train, y_train)

    # Predict using the validation dataset
    y_pred_dt = dt_model.predict(final_X_val)

    # Compute R-squared score
    r2 = r2_score(y_val, y_pred_dt)
    r2_scores.append(r2)

    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_pred_dt))
    rmse_scores.append(rmse)

    print(f'leaf_sample = {leaf_samples} -------> R² = {r2:.4f}, RMSE = {rmse:.4f}')
    print()
```
